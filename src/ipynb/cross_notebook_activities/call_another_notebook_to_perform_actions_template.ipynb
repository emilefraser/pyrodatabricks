## Option 1 - Notebook Workflows using dbutils.notebook.run

* Notebook workflows are a complement to %run because they let you pass parameters to and return values from a notebook. This allows you to build complex workflows and pipelines with dependencies.
* Unlike %run, the dbutils.notebook.run() method starts a new job to run the notebook.
* The methods available in the dbutils.notebook API to build notebook workflows are: run and exit. Both parameters and return values must be strings.

## Option 2 - %run
* Link: https://docs.microsoft.com/en-us/azure/databricks/notebooks/notebooks-use#run
* The %run command allows you to include another notebook within a notebook. 
* You can use %run to modularize your code, for example by putting supporting functions in a separate notebook. You can also use it to concatenate notebooks that implement the steps in an analysis. 
* When you use %run, the called notebook is immediately executed and the functions and variables defined in it become available in the calling notebook.
* The other notebook's code is essentially coped into your current notebook and executed. 

### Recommended Use Cases
1. Call another notebook that contains helper functions to use in your current notebook without manually duplicating code e.g. to mount a directory

### NOTE
1. The run command MUST be the only code in the cell. Even comments can't be there otherwise it wont work. It tries to interprate the reference to a .py file instead of a notebook in that case and cannot find it.
2. This also works when the notebooks are in a Git repo

### Examples

If `notebookA` and `notebookB` are in the same directory, you can use this syntax.

`%run ./notebookA`

If `notebookA` and `notebookB` are not in the same directory, provide the relative path.

`%run ../someDirectory/notebookA`
